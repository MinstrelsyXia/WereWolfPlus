```python
import os
import re

def is_werewolf(player_name, content):
    """
    检查玩家是否是狼人
    """
    return bool(re.search(rf"Werewolf--{player_name}", content))

# 定义关键词列表
# keywords = ["IRP", "KSR", "VSS", "wolves_kpi", "witch_kpi", "seer_kpi", "hunter_kpi", "guard_kpi"]
keywords = ["IRP", "KSR", "VSS", "werewolf_kpi", "seer_kpi", "guard_kpi"]

# 存储结果的字典
# results = {"IRP":[],"KSR":[], "VSS":[], 
#            "wolves_kpi":[], "witch_kpi":[], "seer_kpi":[], "hunter_kpi":[], "guard_kpi":[], 
#            "good_sheriff_kpi":[], "bad_sheriff_kpi":[]}
results = {"IRP":[],"KSR":[], "VSS":[], 
           "werewolf_kpi":[], "seer_kpi":[], "guard_kpi":[], 
           "good_sheriff_kpi":[], "bad_sheriff_kpi":[]}
           

folder_path = r"C:\Users\lenovo\Desktop\博弈论与多智能体\大作业\werewolf\output_xxy\output_my\doubao\werewolf\scene1\doubao_vs_deepseek-v1\\"



# 获取当前目录下所有 .log 文件
log_files = [folder_path+f for f in os.listdir(folder_path) if f.endswith('.log')]
print(log_files)

# 遍历每个 .log 文件
for log_file in log_files:
    with open(log_file, 'r', encoding='utf-8') as file:
        content = file.read()
    
    # 初始化每个文件的结果
    file_results = {}
    
    # 遍历关键词，提取最后出现的数字
    for keyword in keywords:
        matches = re.findall(rf"{keyword}:\s*(-?\d+\.?\d*)?", content)
        if keyword == "werewolf_kpi":
            matches = re.findall(r"werewolf_kpi\s*(-?\d+\.?\d*)?", content)
        if matches:
            file_results[keyword] = matches[-1]  # 取最后一个匹配的数字
            results[keyword].append(float(matches[-1]))
    
    # # 提取射击目标
    # shoot_matches = re.findall(r"'shoot':\s*'(\w+)'", content)
    # for target in shoot_matches:
    #     # print(target)
        
    #     if is_werewolf(target, content):
    #         results["hunter_kpi"].append(1)  # 如果是狼人，记录为1
    #     else:
    #         results["hunter_kpi"].append(0)
    
    # 提取每一轮的decision change指标和警长信息
    round_matches = re.findall(r"Round (\d+) - Decision Change: (-?\d+\.?\d*)", content)
    sheriff_matches = re.findall(r"Round (\d+) - Sheriff: (\w+)", content)
    
    for round_num, decision_change in round_matches:
        # 查找对应轮次的警长
        sheriff = next((sheriff_name for r, sheriff_name in sheriff_matches if r == round_num), None)
        if sheriff:
            if is_werewolf(sheriff, content):
                results["bad_sheriff_kpi"].append(float(decision_change))  # 警长是狼人
            else:
                results["good_sheriff_kpi"].append(float(decision_change))  # 警长是好人
        

# 打印结果
for keyword, values in results.items():
    print(f"{keyword}: {values}")

result_avg= {}
for keyword, values in results.items():
    if values:
        # print(f"{keyword}: {sum(values) / len(values):.2f} (平均值)")
        result_avg[keyword] = sum(values) / len(values)
    else:
        print(f"{keyword}: 无数据")
# result_avg["KSR"]/=2
# result_avg["KRS"]=(result_avg["witch_kpi"] + result_avg["seer_kpi"] + result_avg["hunter_kpi"] + result_avg["guard_kpi"]) / 4
result_avg["KRS"]=( result_avg["seer_kpi"] + result_avg["guard_kpi"]) / 2
result_avg["KRE"]=(result_avg["KRS"]+result_avg["KSR"])/2
print(result_avg)
```
